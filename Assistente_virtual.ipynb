```python
from nltk.util import ngrams

def calcular_similaridade(texto1, texto2):
  # Criar ngrams para ambos os textos
  ngrams1 = set(ngrams(texto1.split(), 2))
  ngrams2 = set(ngrams(texto2.split(), 2))

  # Calcular a similaridade (por exemplo, usando Jaccard similarity)
  return len(ngrams1.intersection(ngrams2)) / len(ngrams1.union(ngrams2))
```
